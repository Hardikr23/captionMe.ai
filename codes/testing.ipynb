```python
def get_named_entity(all_captions):
    import spacy
#     from spacy.en import English

    NER = spacy.load("en_core_web_sm")
    named_entity_list = []
    for caption in all_captions:
        named_entity = NER(caption)
        named_entity_list += list(named_entity.noun_chunks)
    return named_entity_list

all_nouns = get_named_entity(all_captions)
print(len(all_nouns))
```

```python
def pre_process_captions(caption):
    import re
    # Removing Hashtags
    caption = re.sub(r'#.*[)(*&^%$#@!:;<>,.?/"''")]','',caption)
    # Removing Emojis
    caption = caption.encode('ascii', 'ignore').decode('ascii')
    
    return caption
```
